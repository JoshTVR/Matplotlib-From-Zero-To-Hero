## 2.1. Line Plots
Line plots are ideal for showing trends over time or visualizing the relationship between two continuous variables. They connect data points using line segments.

To create a line plot, we use the `plt.plot()` function. We can pass it a list or array of values for the Y-axis, or two lists/arrays for both the X and Y axes.


As another example, this time introducing data into our plot, we’ll create a simple array to represent our data. The `plt.xlabel()` or `plt.ylabel()` functions set the label for the selected axis — whichever we choose. Finally, we add the data to our plot by simply passing the variable we created.

````python
rnum1 = [2,4,6,8,10]
plt.ylabel(rnum1)
plt.grid(True)
plt.plot(rnum1)
````

But what does the official documentation say about this?  
Because if we check the Matplotlib documentation, it can actually be a bit confusing at first, especially when trying to directly find the plotting functionalities.

> The optional parameter *fmt* is a convenient way for defining basic formatting like color, marker, and linestyle. It's a shortcut string notation described in the *Notes* section below.
>
>```python
> plot(x, y)           # plot x and y using default line style and color  
> plot(x, y, 'bo')     # plot x and y using blue circle markers  
> plot(y)              # plot y using x as index array 0..N-1  
> plot(y, 'r+')        # same as above, but with red plus markers  
>```

At this point, what you can see in the documentation is that an **optional parameter** (`fmt`) can be used as a **convenient way** to define the basic formatting of the plot — in essence, it lets you specify *“what goes where”*.

---

#### **What does this mean in simple words?**

The `plt.plot()` function draws a line (or points) on a graph.

* If you pass **only one list/array**, Matplotlib assumes these are **Y values** and automatically uses `0, 1, 2, ...` as **X**.
* If you pass **two lists/arrays**, the first represents **X** and the second **Y**.
* The `fmt` argument is a shortcut to quickly set the **color**, **line style**, and **marker** in a single string.
* You can also use **named parameters** like `color=`, `linestyle=`, `marker=`, etc., for more control.

### 2.x What is `fmt` and Why Does It Exist?

**Brief theoretical background:**  
`fmt` is a **shortcut** for combining **color**, **marker**, and **line style** without having to write multiple separate parameters.  
General format: `"[color][marker][linestyle]"`

* **Short color codes:** `b g r c m y k` (or full names like `'crimson'`, or even `#hex` values).
* **Markers:** `'o'` (circle), `'s'` (square), `'^'` (triangle), etc.
* **Line styles:** `'-'` (solid), `'--'` (dashed), `'-.'` (dash-dot), `':'` (dotted).

````python
rnum1 = [2, 4, 6, 8, 10]

plt.grid(True)
plt.plot(rnum1, 'ro--')  # r = red, o = circle marker, -- = dashed line
plt.title("fmt = 'ro--'")
plt.grid(True)
plt.show()
````

> You can use **fine-grained control** by specifying separate parameters:
> `color=`, `marker=`, `linestyle=`, `markersize=`, `markerfacecolor=`, etc.

### 2.x Atomic Customization: `color`, `marker`, `linestyle`, `markersize`, `markerfacecolor`

**Brief theoretical background:**  
When `fmt` isn’t enough, use **explicit parameters** for better clarity and full control.

**Minimal example (same plot, different styling options):**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(5)
y = np.array([2, 5, 1, 3, 4])

plt.plot(
    x, y,
    color="crimson",          # Line color
    marker="o",               # Marker shape
    linestyle="--",           # Dashed line
    markersize=6,             # Marker size
    markerfacecolor="white",  # Inner marker color
    markeredgecolor="crimson" # Marker border color
)
plt.title("Explicit Style Parameters")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

````python
x = np.arange(5)
y = np.array([2, 5, 1, 3, 4])

plt.plot(
    x, y,
    color="Crimson",
    marker="o",
    linestyle="--",
    markersize=6,
    markerfacecolor="white",
    markeredgecolor="crimson"
)
plt.title("Parámetros explícitos de estilo")
plt.xlabel("x"); plt.ylabel("y")
plt.grid(True)
plt.show()
````

### 2.x `title`, `xlabel`, `ylabel`, `grid`, `show`: What They Do and When to Use Them

* **`plt.title("...")`**: Provides context for the plot with a short, descriptive title.
* **`plt.xlabel("...")` / `plt.ylabel("...")`**: Clearly indicate **what each axis represents**.
* **`plt.grid(True)`**: Helps **read the values** more easily; you can customize it with options like `axis='x'/'y'/'both'`, `linestyle=':'`, etc.
* **`plt.show()`**: In Jupyter notebooks, plots are often displayed automatically, but in **scripts** it ensures that the plot is **rendered exactly at that point**.

**Minimal example:**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(5)
y = x**2

plt.plot(x, y)
plt.title("Simple Square")
plt.xlabel("x")
plt.ylabel("x^2")
plt.grid(True)
plt.show()
```

### 2.x Combining Pandas and Line Plots (Keeping It Simple)

**Brief theoretical background:**  
Pandas can plot using Matplotlib under the hood.  
If you have a **Series** (a column with an ordered index), you can easily plot it **directly as a line**.

**Minimal example using your Pokémon dataset (average Attack by Stage):**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Pokemon.csv")
attack_per_stage = df.groupby('Stage')['Attack'].mean().sort_index()

attack_per_stage.plot(kind='line', marker='o')
plt.title("Average Attack per Stage")
plt.xlabel("Stage")
plt.ylabel("Average Attack")
plt.grid(True)
plt.show()

````python
# Ask the user to enter Pokédex numbers
entrada = input("Enter Pokédex numbers separated by commas (e.g., 6,25,94): ")

# Convert the input into a list of integers
try:
    numeros = list(map(int, entrada.split(',')))
except ValueError:
    print("Error: you must enter only numbers separated by commas.")
    exit()

# Filter Pokémon by those Pokédex numbers
selected_pokemon = pokes[pokes['Num'].isin(numeros)]

# Validate if any Pokémon were found
if selected_pokemon.empty:
    print("No Pokémon found with those numbers.")
    exit()

# Select base stats columns
stats_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
stats_data = selected_pokemon[stats_columns]
names = selected_pokemon['Name'].values

# Display selected Pokémon names and their stats
print(selected_pokemon['Name'])
print(stats_data)
````

### Why Use Line Plots to Visualize Statistics?

Line plots are useful when you want to compare how multiple values vary along the same axis.  
In this case:

- The **X-axis** represents the different statistics (`HP`, `Attack`, etc.).
- The **Y-axis** represents the numerical value of each statistic.

This allows us to easily see which Pokémon is faster, which one has better defense, and so on.  
The lines help us follow the **shape** of each Pokémon’s statistical profile, making visual comparisons between them much easier.


````python
# Transpose to make visualization easier
stats_data_t = stats_data.T
stats_data_t.columns = selected_pokemon['Name']
stats_data_t
````

````python
import matplotlib.pyplot as plt

# Create the line plot
plt.figure(figsize=(20, 10))
for name in stats_data_t.columns:
    values = stats_data_t[name].values
    plt.plot(stats_data_t.index, stats_data_t[name], marker='o', label=name)
    
    # Add text at the end of the line (last data point)
    total = values.sum()
    x = len(stats_data_t.index) - 1  # Last position on the X-axis
    y = values[-1]                   # Y-value at the last point
    plt.text(x + 0.1, y, f'Total: {total}', fontsize=9, va='center')

plt.title('Base Stats of the First 5 Pokémon')
plt.xlabel('Statistic')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

## Line-by-line Explanation

  * A large size improves readability when there are many data series.

---

```python

* `stats_data_t` is likely a **transposed DataFrame** (`.T`), where:

  * **Rows** (`index`) represent stat types: `HP`, `Attack`, `Defense`, etc.
  * **Columns** (`columns`) are Pokémon or categories.

---

```python

---

```python

---

---
