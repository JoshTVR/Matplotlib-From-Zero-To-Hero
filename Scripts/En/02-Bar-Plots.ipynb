## 2.3. Bar Plots

Bar plots are useful for comparing quantities across different categories. Each bar represents a category, and its height (or length) indicates the value of that category.

The `plt.bar()` function is used for vertical bar plots, and `plt.barh()` for horizontal bar plots

## What it is and when to use it

* **What it is:** bars to compare **quantities** across **categories**.
* **When to use it:** **categorical** X-axis (types/groups) and **numeric** Y-axis (count, average, sum).
* **Avoid:** too many categories or very long labels (consider Top-N or horizontal bars).

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

categorias = ["A", "B", "C"]
valores = [10, 7, 12]

plt.bar(categorias, valores)   # vertical por defecto
plt.title("Barras básicas")
plt.xlabel("Categoría")
plt.ylabel("Valor")
plt.grid(axis='y', linestyle=':', alpha=0.6)  # rejilla solo en Y
plt.show()
```

```python
# Datos de ejemplo
categorias = ["A", "B", "C", "D"]
valores = [23, 45, 56, 12]

colores = ["red", "green", "blue", "purple"]

# Crear un gráfico de barras
plt.bar(categorias, valores, color=colores)
plt.title("Gráfico de Barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

```python
plt.bar(
    categorias, valores,
    color="steelblue",
    edgecolor="black",
    linewidth=1,
    alpha=0.9
)
plt.title("Personalización de barras")
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.show()
```

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, color="orchid")
plt.title("Germination Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="red")
plt.title("Failure Rates")
plt.ylabel("Number Of Dead Seedling")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db")
plt.bar(plants, germinated, width=0.4, color="#2ecc71", align="edge")
```

```python
plt.figure(figsize=(6,6))
plt.barh(plants, germinated, color="#2ecc71")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db", bottom=germinated )
plt.bar(plants, germinated, color="#2ecc71")
```
