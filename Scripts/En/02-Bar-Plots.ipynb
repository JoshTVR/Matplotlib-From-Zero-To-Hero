## 2.3. Bar Plots

Bar plots are useful for comparing quantities across different categories. Each bar represents a category, and its height (or length) indicates the value of that category.

The `plt.bar()` function is used for vertical bar plots, and `plt.barh()` for horizontal bar plots

## What it is and when to use it

* **What it is:** bars to compare **quantities** across **categories**.
* **When to use it:** **categorical** X-axis (types/groups) and **numeric** Y-axis (count, average, sum).
* **Avoid:** too many categories or very long labels (consider Top-N or horizontal bars).

```python
import matplotlib.pyplot as plt
```

## 1. Basic Bar Plots

```python

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Vertical bar plot
plt.bar(categories, values)
plt.title("Basic Vertical Bar Plot")
plt.show()

# Horizontal bar plot
plt.barh(categories, values)
plt.title("Basic Horizontal Bar Plot")
plt.show()
```

## 2. Customizing Bars

```python

import numpy as np

x = np.arange(len(categories))

# Custom colors, width, alpha, edge color
plt.bar(x, values, color=['skyblue','salmon','lightgreen','orange'],
        width=0.6, alpha=0.8, edgecolor='black', linewidth=1.5)
plt.xticks(x, categories)
plt.title("Customized Bars")
plt.show()
```

## 3. Labels, Ticks & Grid

```python

plt.bar(categories, values, color='steelblue')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot with Labels & Grid")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 4. Error Bars

```python

errors = [0.5, 0.7, 0.2, 0.4]

plt.bar(categories, values, yerr=errors, capsize=5, color='lightcoral')
plt.title("Bar Plot with Error Bars")
plt.show()
```

## 5. Grouped Bar Charts

```python

values2 = [6, 3, 5, 2]
width = 0.35

plt.bar(x - width/2, values, width, label='Group 1')
plt.bar(x + width/2, values2, width, label='Group 2')
plt.xticks(x, categories)
plt.title("Grouped Bar Chart")
plt.legend()
plt.show()
```

## 6. Stacked Bar Charts

```python

plt.bar(categories, values, label='Group 1')
plt.bar(categories, values2, bottom=values, label='Group 2')
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

## 7. Annotating Bars

```python

fig, ax = plt.subplots()
bars = ax.bar(categories, values, color='teal')
ax.set_title("Bar Plot with Annotations")

# Add labels on top of bars
ax.bar_label(bars, fmt='%d', padding=3)
plt.show()
```

## 8. Styling & Themes

```python

plt.style.use('ggplot')
plt.bar(categories, values, color='purple')
plt.title("Bar Plot with ggplot Style")
plt.show()

plt.style.use('dark_background')
plt.bar(categories, values, color='gold')
plt.title("Bar Plot with Dark Background")
plt.show()

# Reset style
plt.style.use('default')
```

## 9. Advanced Bar Plots

```python

# Log scale example
plt.bar(categories, [1, 10, 100, 1000])
plt.yscale('log')
plt.title("Logarithmic Scale Bar Plot")
plt.show()
```

```python

# Polar bar plot
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
theta = np.linspace(0.0, 2*np.pi, len(values), endpoint=False)
bars = ax.bar(theta, values, width=0.5, color=plt.cm.viridis(values))
ax.set_title("Polar Bar Plot", va='bottom')
plt.show()
```

```python

# 3D Bar Plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

_x = np.arange(len(categories))
_y = np.array([0,1,2,3])
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()
top = np.random.randint(1, 10, size=len(x))
bottom = np.zeros_like(top)
width = depth = 0.4

ax.bar3d(x, y, bottom, width, depth, top, shade=True)
ax.set_title("3D Bar Plot")
plt.show()
```

## 10. Bar Plots with Pandas

```python

import pandas as pd

data = {
    'Group1': values,
    'Group2': values2
}
df = pd.DataFrame(data, index=categories)

# Simple bar
df.plot.bar(title="Pandas Bar Plot")
plt.show()

# Stacked bar
df.plot.bar(stacked=True, title="Pandas Stacked Bar Plot")
plt.show()
```

## 11. Comparison with Seaborn

```python

import seaborn as sns

sns.barplot(x=categories, y=values, palette='viridis')
plt.title("Seaborn Bar Plot")
plt.show()
```
