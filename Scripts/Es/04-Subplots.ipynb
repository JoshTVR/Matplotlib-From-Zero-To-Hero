# 5.3. Diseños de Gráficos Complejos con `GridSpec`

Para diseños de subplots más complejos que no encajan en una cuadrícula regular, Matplotlib ofrece `GridSpec`. `GridSpec` permite especificar la geometría de la cuadrícula en la que se colocará un subplot, y luego colocar subplots que abarquen varias filas o columnas.

## Uso Básico de `GridSpec`
Primero, creamos una instancia de `GridSpec` y luego usamos `add_subplot` con la notación de slicing para especificar dónde se ubicará cada subplot.


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(10, 8))
gs = GridSpec(3, 3) # Cuadrícula de 3 filas por 3 columnas

ax1 = fig.add_subplot(gs[0, :]) # Primer subplot ocupa toda la primera fila
ax1.plot(np.random.rand(100), color="red")
ax1.set_title("Gráfico 1 (Fila Completa)")

ax2 = fig.add_subplot(gs[1, :-1]) # Segundo subplot ocupa las dos primeras columnas de la segunda fila
ax2.plot(np.random.rand(100), color="green")
ax2.set_title("Gráfico 2 (2/3 de Fila)")

ax3 = fig.add_subplot(gs[1:, -1]) # Tercer subplot ocupa la última columna de las filas 1 y 2
ax3.plot(np.random.rand(100), color="blue")
ax3.set_title("Gráfico 3 (2 Filas, 1 Columna)")

ax4 = fig.add_subplot(gs[2, 0]) # Cuarto subplot en la última fila, primera columna
ax4.plot(np.random.rand(100), color="purple")
ax4.set_title("Gráfico 4")

ax5 = fig.add_subplot(gs[2, 1]) # Quinto subplot en la última fila, segunda columna
ax5.plot(np.random.rand(100), color="orange")
ax5.set_title("Gráfico 5")

plt.tight_layout()
plt.show()
```
