## Gráficos de Tarta (Pie Charts)
Los gráficos de tarta son útiles para mostrar la proporción de diferentes categorías dentro de un todo. Cada 'rebanada' representa una categoría y su tamaño es proporcional a su valor.

Se crean con `plt.pie()`.


```python
labels = ["Turkey", "Potatoes", "Pumpkin Pie", "Stuffing"]
prices = [25.99,3.24,9.50, 6.99]
```

```python
plt.pie(prices, labels=labels, autopct="%1.1f%%", shadow=True, explode=(0,0,0.1,0))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
sizes = [15, 30, 45, 10] # Proporciones de cada categoría
labels = ["Manzanas", "Naranjas", "Plátanos", "Uvas"]
colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode = (0, 0.1, 0, 0)  # Solo 'explotar' la segunda rebanada (Naranjas)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")  # Asegura que el círculo sea un círculo.
plt.title("Distribución de Frutas")
plt.show()
```
